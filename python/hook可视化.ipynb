```python
import torch
import torchvision.models as models
from PIL import Image
from torchvision import transforms as T
import matplotlib.pyplot as plt
from Net import Cnn_250, AlexNet_250, Dnn_250, Bilinear_250, Cnn_150, AlexNet_150, Dnn_150, Bilinear_150, Bilinear_150_bnrelu

device = 'cpu'

state_dict = torch.load('./save_model/Cnn_150_batch_16_epochs_100_input_150best.pth')
model = Cnn_150().to(device)
model.load_state_dict(state_dict)
model.eval()

# 定义hook
class SaveOutput:
    def __init__(self):
        self.outputs = []

    def __call__(self, module, module_in, module_out):
        self.outputs.append(module_out)

    def clear(self):
        self.outputs = []

# 对Conv2d注册hook
save_output = SaveOutput()
hook_handles = []
for layer in model.modules():
    if isinstance(layer, torch.nn.modules.conv.Conv2d):
        handle = layer.register_forward_hook(save_output)
        hook_handles.append(handle)


image = Image.open('pict.jpg')
transform = T.Compose([T.Resize((150, 150)), T.ToTensor()])
X = transform(image).unsqueeze(dim=0).to(device)

out = model(X)
def module_output_to_numpy(tensor):
    return tensor.detach().to('cpu').numpy()

images = module_output_to_numpy(save_output.outputs[0])

with plt.style.context("seaborn-white"):
    plt.figure(figsize=(10, 10), frameon=False)
    for idx in range(16):
        plt.subplot(4, 4, idx+1)
        plt.imshow(images[0, idx])
    plt.setp(plt.gcf().get_axes(), xticks=[], yticks=[]);
    plt.show()
```

```python
images = module_output_to_numpy(save_output.outputs[-1])

with plt.style.context("seaborn-white"):
    plt.figure(figsize=(10, 10), frameon=False)
    for idx in range(16):
        plt.subplot(4, 4, idx+1)
        plt.imshow(images[0, idx])
    plt.setp(plt.gcf().get_axes(), xticks=[], yticks=[]);
    plt.show()
```
